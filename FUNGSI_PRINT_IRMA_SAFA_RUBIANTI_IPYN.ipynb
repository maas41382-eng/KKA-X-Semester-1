### 1. Multiple Assignment (Tuple Unpacking)

```python
# Anda dapat menetapkan beberapa variabel pada satu baris dengan memberikan beberapa nilai yang dipisahkan koma.
a, b, c = 10, 20, 30
print(f"a: {a}, b: {b}, c: {c}")

# Ini juga bekerja dengan tipe data yang berbeda.
x, y, z = "Apple", 3.14, True
print(f"x: {x}, y: {y}, z: {z}")
```

### 2. Menetapkan Nilai yang Sama ke Beberapa Variabel

```python
# Anda dapat menetapkan nilai tunggal ke beberapa variabel secara bersamaan.
p = q = r = 100
print(f"p: {p}, q: {q}, r: {r}")
```

### 3. Menggunakan List atau Tuple untuk Unpacking

```python
# Anda dapat membongkar (unpack) elemen dari list atau tuple ke beberapa variabel.
my_list = [1, 2]
var1, var2 = my_list
print(f"var1: {var1}, var2: {var2}")

my_tuple = ('hello', 'world', 123)
word1, word2, number = my_tuple
print(f"word1: {word1}, word2: {word2}, number: {number}")
```
