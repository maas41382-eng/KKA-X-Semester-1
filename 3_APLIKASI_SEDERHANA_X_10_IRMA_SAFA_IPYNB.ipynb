#aplikasi menghitung luas persegi panjang

```python
def hitung_luas_persegi_panjang(panjang, lebar):
    """
    Menghitung luas persegi panjang.

    Args:
        panjang (int): Panjang dari persegi panjang.
        lebar (int): Lebar dari persegi panjang.

    Returns:
        int: Luas dari persegi panjang.
    """
    luas = panjang * lebar
    return luas

# Prompt the user to enter the length (panjang) of the rectangle.
panjang_input = int(input("Masukkan panjang dari persegi panjang (cm): "))

# Prompt the user to enter the width (lebar) of the rectangle.
lebar_input = int(input("Masukkan lebar dari persegi panjang (cm): "))

# Calculate the area (luas) by calling the function.
luas_hasil = hitung_luas_persegi_panjang(panjang_input, lebar_input)

# Print the calculated area to the console.
print(f"\nLuas dari persegi panjangnya adalah {luas_hasil} cm^2.")
```
