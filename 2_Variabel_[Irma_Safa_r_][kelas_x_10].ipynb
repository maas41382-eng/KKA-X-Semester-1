### Declaring Variables Correctly

Python variable names must start with a letter or an underscore and can only contain alpha-numeric characters and underscores. They are also case-sensitive.

```python
# Valid variable names
my_variable = "Hello"
user_age = 30
_internal_value = 100

print(my_variable)
print(user_age)
print(_internal_value)
```

### Common Mistakes to Avoid

```python
# Invalid: Cannot start with a number
# 1st_name = "John"
# print(1st_name)

# Invalid: Cannot contain spaces
# my name = "Alice"
# print(my name)

# Invalid: Cannot use Python keywords as variable names (e.g., 'for', 'print', 'if')
# for = "loop_variable"
# print(for)

# Python is case-sensitive
myVariable = "Value One"
myvariable = "Value Two"

print(myVariable)
print(myvariable)
```

### Types of Assignments

```python
# Single Assignment
student_name = "Budi"
student_grade = 85
print(f"Student: {student_name}, Grade: {student_grade}")

# Multiple Assignment
x, y, z = 10, 20, 30
print(f"x: {x}, y: {y}, z: {z}")

# Chain Assignment
a = b = c = "Python"
print(f"a: {a}, b: {b}, c: {c}")
```
